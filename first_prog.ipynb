**Task 2**

Write a function that takes a two-dimensional array and a string as input and returns an array rotated 90 degrees counterclockwise if the string 'left' was passed, and clockwise if the string 'right' was passed.

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Do not use numpy and nested loops. Use list comprehensions and built-in function `zip`.*Курсив*

````python
def rotate_array(matrix, direction):
    return [list(row) for row in zip(*[row[::-1] for row in matrix])] if direction == 'right' else [list(row)[::-1] for row in zip(*matrix)]


matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(rotate_array(matrix, 'left'))
print(rotate_array(matrix, 'right'))
````

### Implementing a Library Management System

#### Description

You are required to design and implement a system for managing books and users in a library. The system should allow for the management of books (adding, deleting, searching by various criteria) and users (registration, deletion, searching), as well as tracking the history of interactions between them (issuing and returning books).

#### Tasks

1. **`Book` Class**:
   - Attributes: title, author, year of publication, ISBN, number of copies.
   - Methods: constructor, methods to get information about the book, method to change the number of copies (when issuing and returning books).

2. **`User` Class**:
   - Attributes: user name, library card number, list of borrowed books.
   - Methods: constructor, methods for user registration, methods for adding and removing books from the borrowed list.

3. **`Library` Class**:
   - Attributes: list of books, list of users, transaction history (who, when, which book was borrowed and returned).
   - Methods: constructor, methods for adding and deleting books and users, methods for issuing and returning books, searching for books and users by various criteria, method to display the transaction history.

#### Assignment

1. Implement the `Book`, `User`, and `Library` classes with the specified attributes and methods.
2. Create several books and users, and add them to the library system.
3. Implement scenarios for issuing books to users and their return.
4. Display the transaction history to show how books were issued and returned.


````python
class Book:
    def __init__(self, title, author, year, isbn, number_of_copies):
        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.number_of_copies = number_of_copies
        print(f"Book {title} created")

    def get_info(self):
        return f"<{self.title}> {self.author}, {self.year}, {self.isbn}, copies: {self.number_of_copies}"

    def change_number(self, delta):
        if self.number_of_copies + delta < 0:
            return False
        self.number_of_copies += delta
        return True


class User:
    def __init__(self, name, card_number):
        self.name = name
        self.card_number = card_number
        self.list_books = []
        print(f"User {name} created")

    def add_book_to_list(self, book):
        if book in self.list_books:
            return False
        self.list_books.append(book)
        return True

    def remove_book_from_list(self, book):
        if book not in self.list_books:
            return False
        self.list_books.remove(book)
        return True

    def list_borrowed_books(self):
        return [b.get_info() for b in self.list_books]


class Library:
    def __init__(self):
        self.books = []
        self.users = []
        self.transaction_history = []
        print("Library created")

    def add_book(self, book):
        if self.find_book_by_isbn(book.isbn) is not None:
            return False
        self.books.append(book)
        return True

    def delete_book_by_isbn(self, isbn):
        book = self.find_book_by_isbn(isbn)
        if book is None:
            return False
        for u in self.users:
            for b in u.list_books:
                if b.isbn == isbn:
                    return False
        self.books.remove(book)
        return True

    def register_user(self, user):
        if self.find_user_by_card(user.card_number) is not None:
            return False
        self.users.append(user)
        return True

    def delete_user_by_card(self, card_number):
        user = self.find_user_by_card(card_number)
        if user is None:
            return False
        if user.list_books:
            return False
        self.users.remove(user)
        return True

    def issue_books(self, card_number, isbn):
        user = self.find_user_by_card(card_number)
        book = self.find_book_by_isbn(isbn)
        if user is None or book is None:
            return False
        if not book.change_number(-1):
            return False
        if not user.add_book_to_list(book):
            book.change_number(+1)
            return False
        self.transaction_history.append(f"{user.name} issued {book.title}")
        return True

    def return_books(self, card_number, isbn):
        user = self.find_user_by_card(card_number)
        book = self.find_book_by_isbn(isbn)
        if user is None or book is None:
            return False
        if not user.remove_book_from_list(book):
            return False
        book.change_number(+1)
        self.transaction_history.append(f"{user.name} returned {book.title}")
        return True

    def search_book(self, title=None, author=None, year=None, isbn=None):
        result = []
        for b in self.books:
            ok = True
            if title is not None and title.lower() not in b.title.lower():
                ok = False
            if author is not None and author.lower() not in b.author.lower():
                ok = False
            if year is not None and year != b.year:
                ok = False
            if isbn is not None and isbn != b.isbn:
                ok = False
            if ok:
                result.append(b.get_info())
        return result

    def search_user(self, name=None, card_number=None):
        result = []
        for u in self.users:
            ok = True
            if name is not None and name.lower() not in u.name.lower():
                ok = False
            if card_number is not None and card_number != u.card_number:
                ok = False
            if ok:
                result.append(f"{u.name}, card: {u.card_number}, borrowed: {len(u.list_books)}")
        return result

    def display_transaction(self):
        for record in self.transaction_history:
            print(record)

    def find_book_by_isbn(self, isbn):
        for b in self.books:
            if b.isbn == isbn:
                return b
        return None

    def find_user_by_card(self, card_number):
        for u in self.users:
            if u.card_number == card_number:
                return u
        return None


````

````python
# создаём библиотеку
library = Library()


# создаём 10 классических зарубежных книг
b1 = Book("Гордость и предубеждение", "Джейн Остин", 1813, "9780141439518", 3)
b2 = Book("1984", "Джордж Оруэлл", 1949, "9780451524935", 2)
b3 = Book("Убить пересмешника", "Харпер Ли", 1960, "9780061120084", 4)
b4 = Book("Великий Гэтсби", "Фрэнсис Скотт Фицджеральд", 1925, "9780743273565", 2)
b5 = Book("Сто лет одиночества", "Габриэль Гарсиа Маркес", 1967, "9780060883287", 5)
b6 = Book("Моби Дик", "Герман Мелвилл", 1851, "9780142437247", 3)
b7 = Book("Над пропастью во ржи", "Джером Сэлинджер", 1951, "9780316769488", 2)
b8 = Book("Три товарища", "Эрих Мария Ремарк", 1936, "9780099532811", 2)
b9 = Book("Портрет Дориана Грея", "Оскар Уайльд", 1890, "9780141442464", 3)
b10 = Book("Собор Парижской Богоматери", "Виктор Гюго", 1831, "9780451527882", 4)


# добавляем книги в библиотеку
print("\nДобавляем книги в библиотеку")
library.add_book(b1)
library.add_book(b2)
library.add_book(b3)
library.add_book(b4)
library.add_book(b5)
library.add_book(b6)
library.add_book(b7)
library.add_book(b8)
library.add_book(b9)
library.add_book(b10)


# создаём пользователей
u1 = User("Синельникова Юлия", "U2001")
u2 = User("Иванов Иван", "U2002")
u3 = User("Никонов Николай", "U2003")
u4 = User("Синельников Кирилл", "U2004")
u5 = User("Петров Петр", "U2005")


# регистрируем пользователей
print("\nРегистрируем пользователей")
library.register_user(u1)
library.register_user(u2)
library.register_user(u3)
library.register_user(u4)
library.register_user(u5)


# выдаём книги пользователям
print("\nВыдаём книги пользователям")
library.issue_books("U2001", "9780141439518")  # Гордость и предубеждение
library.issue_books("U2001", "9780451524935")  # 1984
library.issue_books("U2002", "9780061120084")  # Убить пересмешника
library.issue_books("U2003", "9780060883287")  # Сто лет одиночества
library.issue_books("U2004", "9780142437247")  # Моби Дик
library.issue_books("U2005", "9780099532811")  # Три товарища


# пробуем выдать книгу, экземпляры которой закончились
print("\nПробуем выдать книгу без свободных экземпляров")
library.issue_books("U2003", "9780451524935")
library.issue_books("U2004", "9780451524935")


# возвращаем книгу
print("\nВозвращаем книгу")
library.return_books("U2001", "9780451524935")


# снова выдаём её другому пользователю
print("\nСнова выдаём освобождённую книгу")
library.issue_books("U2003", "9780451524935")


# ищем книги Оруэлла
print("\nПоиск книг Оруэлла")
print(library.search_book(author="Оруэлл"))


# ищем пользователя по имени
print("\nПоиск пользователя по имени Юлия")
print(library.search_user(name="Юлия"))


# пытаемся удалить книгу, которая выдана
print("\nПопытка удалить выданную книгу")
library.delete_book_by_isbn("9780141439518")


# возвращаем книгу и удаляем её
print("\nВозвращаем книгу и удаляем её")
library.return_books("U2001", "9780141439518")
library.delete_book_by_isbn("9780141439518")


# пытаемся удалить пользователя с книгами
print("\nПопытка удалить пользователя с книгами")
library.delete_user_by_card("U2003")


# выводим историю транзакций
print("\nИстория операций")
library.display_transaction()
print("\n")
````

**Task 6**

Let
$$A = \sum_{i=1}^{10000} \frac{1}{i^2},\quad B=\sum_{i=10000}^{1} \frac{1}{i^2}.$$
Calculate the values of $A$ and $B$ and compare them. What do you observe? Explain why this happens. What is the best way to calculate the value of $\sum\limits_{i=1}^{10000} \dfrac{1}{i^2}$?

````python
import math

def compute_sums(n):
    # Прямой порядок: от 1 до n
    A = 0.0
    for i in range(1, n + 1):
        A += 1 / (i * i)

    # Обратный порядок: от n до 1
    B = 0.0
    for i in range(n, 0, -1):
        B += 1 / (i * i)

    return A, B

n = 10000
A, B = compute_sums(n)

# Точное значение (приближённое) для сравнения
exact = math.pi**2 / 6 - 1/n

print(f"A (прямой порядок) = {A:.15f}")
print(f"B (обратный порядок) = {B:.15f}")
print(f"Разница A - B = {A - B:.15e}")
print(f"Ошибка A: {abs(A - exact):.15e}")
print(f"Ошибка B: {abs(B - exact):.15e}")


````

При вычислении на компьютере A и B немного отличаются из-за ошибок округления при сложении чисел с плавающей точкой.
B (суммирование от i 10000 до 1) точнее, потому что сначала складываются малые члены, и они меньше теряются на фоне растущей суммы.

Лучший способ: суммировать от меньших членов к большим (т.е. в обратном порядке)
даёт более точный результат из-за особенностей с плавающей точкой.
